```python
from matplotlib import pyplot as plt
import gluoncv
from gluoncv import model_zoo, data, utils
```

```python
net = model_zoo.get_model('yolo3_darknet53_coco', pretrained=True) #importing model
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
x, orig_img = data.transforms.presets.rcnn.load_test('/content/drive/MyDrive/Colab Notebooks/downloadYolo.jpg') #loading image
```

```python
box_ids, scores, bboxes = net(x)# model detcting objects    
ax = utils.viz.plot_bbox(orig_img, bboxes[0], scores[0], box_ids[0], class_names=net.classes) #image with bounding boxes
```
