```python
from matplotlib import pyplot as plt
import gluoncv
from gluoncv import model_zoo, data, utils
```

```python
net = model_zoo.get_model('faster_rcnn_resnet50_v1b_voc', pretrained=True) #importing pre-trained model
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

# New Section

```python
x, orig_img = data.transforms.presets.rcnn.load_test('download.jpg') #loading image
```

```python
box_ids, scores, bboxes = net(x) #detecting objects
ax = utils.viz.plot_bbox(orig_img, bboxes[0], scores[0], box_ids[0], class_names=net.classes)
```

```python
plt.show() #plotting the image with bounding box
```
